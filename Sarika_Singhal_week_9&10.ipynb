```python
#Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import urllib.request, urllib.parse, urllib.error

import requests

from bs4 import BeautifulSoup

import ssl

import re

#Checking SSL certificate


ct = ssl.create_default_context()

ct.check_hostname = False

ct.verify_mode = ssl.CERT_NONE
```

```python
#3 - Read HTML from URL

url = 'https://www.gutenberg.org/browse/scores/top'

status = requests.get(url)
```

```python
#4 - Function to check the status of web request

def request_status_check(r):
    if r.status_code==200:
        print('Success!')
        
    else:
        print('Failure!')
   
request_status_check(status)
```

```python
#5 - Decoding the response and pass this on to BeautifulSoup for HTML parsing

contents = status.content.decode(status.encoding)
soup = BeautifulSoup(contents, 'html.parser')
```

```python
#6 - Find all Href tags
list_link = []
for link in soup.findAll('a'):
    list_link.append(link.get('href'))
print(list_link[:30])
```

```python
#7 - Regular expression

for i in range(1,100):
    link=list_link[i]
    link=link.strip()
    #print(link)
# Regular expression to find the numeric digits in the link (href) string
    n=re.findall('[0-9]+',link)
    if len(n)==1:
        # Append the filenumber casted as integer
        print(n)
```

```python
#9 - What does the soup object's text look like
print(soup.text[:2000])
```

```python
# 10 - Search in the entered text
list_title_temp=[]
start_idx=soup.text.splitlines().index('Top 100 EBooks yesterday')

for i in range(100):
    list_title_temp.append(soup.text.splitlines()[start_idx+9+i])

#printing titles
#print(list_title_temp)

list_titles=[]

for i in range(100):
    id1,id2=re.match('^[a-zA-Z ]*',list_title_temp[i]).span()
    list_titles.append(list_title_temp[i][id1:id2])

for t in list_titles:
    print(t)
```

```python
# 4 - open APIKey file
with open('APIkey.json') as f:
    keys = json.load(f)
    omdbapi = keys['OMDBapi']
```

```python
# 5 - Assingment
url_data = 'http://www.omdbapi.com/?'
```

```python
# 6 - creating a variable called apikey
apikey = '&apikey='+omdbapi
```

```python

#7 - Writing a utility function called print_json to print the movie data from a JSON file

def print_json(json_file):
    print(json_file)
```

```python
#9 -  utility function for searching
def search_movie(title):
    try:
        url = url_data + urllib.parse.urlencode({'t': str(title)})+apikey
        print(url)
        uh = urllib.request.urlopen(url)
        data = uh.read()
        json_data=json.loads(data)
        if json_data['Response']=='True':
            print_json(json_data)


    except urllib.error.URLError as e:
        print(f'ERROR: {e.reason}')
```

```python

# 10 - Titanic pull
movie = 'Titanic'
search_movie(movie)
```
