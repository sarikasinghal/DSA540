```python
#Chapter 11
#Convert between string and date time
#Generate date range
#Frequencies and date offsets
#Convert timestamps to periods and back
#Period Frequency conversions
```

```python
# Generating daterange
DateData = candyData.copy()
date_range = pd.date_range(start='2020-01-01', periods = 309, freq ='BM')
DateData.insert(2,'GeneratedDate',date_range)
DateData.head()
```

```python
# Create the DateOffset
do = pd.tseries.offsets.DateOffset(n = 2)
  
# Print the Timestamp
print(DateData.GeneratedDate)
  
# Print the DateOffset
print(do)
```

```python

# Adding the dateoffset to the given timestamp
new_timestamp = DateData.GeneratedDate + do
  
# Print the updated timestamp
print(new_timestamp)
```
